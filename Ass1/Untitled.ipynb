```python
data = pd.read_csv('25.csv')
```

```python
data[(data['Protocol']=='TCP') | (data['Protocol']=='UDP') |(data['Protocol']=='DNS')]
```

```python
data['floored_time'] = data['Time'].apply(math.floor
                                         )
```

```python
udp=data[data['Protocol']=='UDP'][['Length','floored_time']].groupby('floored_time').sum('length')
```

```python
tcp = data[data['Protocol']=='TCP'][['Length','floored_time']].groupby('floored_time').sum('length')
```

```python
dns = data[data['Protocol']=='DNS'][['Length','floored_time']].groupby('floored_time').sum('length')
```

```python
import matplotlib.pyplot as plt
plt.figure(6)
plt.plot(udp,label='udp')
plt.plot(tcp,label='tcp')
plt.plot(dns,label='dns')

plt.legend()
```
